```python
st = pd.HDFStore("/Users/nipun/Downloads/wiki-all.h5")
```

```python
from collections import Counter
c = Counter()
t = len(st.keys())
homes = []
for i, k in enumerate(st.keys()[:]):
    print(k, i*100//t)
    col = st[k].columns
    if 'waterheater1' in col:
        homes.append(k)
    c.update(col)
    
    
```

```python
pd.Series(c).sort_values(ascending=False)
```

```python
%matplotlib inline
```

```python
o = {}
for home in homes:
    try:
        print(home)

        d = st[home]['waterheater1']['2014']
        o[home]=d.groupby(d.index.hour).mean()
    except:
        pass
```

```python
import seaborn as sns
sns.heatmap(pd.DataFrame(o).T.dropna())
```
